### Alguns significados

**EDT:** Eau de Toilette, que em francês significa "água de toalete". É uma categoria de perfume que geralmente contém entre 5% a 15% de concentração da essência.

**EDP:** Eau de Parfum, que também em francês significa "água de perfume". É uma categoria de perfume que geralmente contém entre 15% a 20% de concentração da essência.

EDT e EDP também podem apresentar diferenças na composição dos ingredientes, no modo como são produzidos e nas técnicas de aplicação.

Geralmente, o EDT é considerado mais leve e adequado para o uso diário, especialmente durante o dia, enquanto o EDP é mais intenso e pode ser mais adequado para ocasiões especiais ou uso noturno.

No entanto, a fixação e a projeção do perfume também dependem de outros fatores, como a química da pele da pessoa, a umidade do ar e a temperatura. Portanto, é importante experimentar as duas opções antes de escolher um perfume e considerar a sua própria química corporal antes de fazer a escolha final.

Aqui na base de dados utilizada nesse projeto existem outras categorias de perfume além do EDT e EDP.

**PdT:** Parfum de Toilette, é uma categoria de perfume que contém uma concentração de essência entre 10% a 20%, geralmente mais forte que o EDT, mas mais suave que o EDP. É menos comum do que as outras categorias de perfume.

**Óleo:** também conhecido como perfume em óleo, é uma forma mais concentrada de fragrância, geralmente com uma concentração de essência de 15% a 30%. É comumente vendido em pequenos frascos roll-on ou conta-gotas.

**EDC:** Eau de Cologne, é uma categoria de perfume que contém uma concentração de essência entre 2% a 5%, geralmente menos concentrada que o EDT. A EDC é conhecida por ser leve e refrescante, e é frequentemente usada como um perfume de verão ou fragrância para depois do banho.

***Mas para esse estudo será utilizado apenas o EDT e EDP, pois a quantidade de ocorrencia para os outros tipos é muito baixa.***
